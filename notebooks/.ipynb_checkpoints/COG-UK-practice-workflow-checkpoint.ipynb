Extract just the leaf tip names from the alignment file, formatting them as they appear on the tree:

```python
readfile = '../data/COG-UK-NORW/NORW_metadata.csv'
df = pd.read_csv(readfile, header=0, usecols=['sequence_name', 'sample_date'])
df = df.rename(columns={'sequence_name':'name', 'sample_date':'date'}) #headers required by treetime

#include only the names that match the names in the alignment
leaf_df = pd. read_csv('../data/COG-UK-NORW/NORW_200_leaftips.csv', header=None)
leaf_df = leaf_df.rename(columns={0:'name'})

#add dates to leaf_df, and format the names like in the tree
save_df = leaf_df.merge(df, how='left')
save_df.loc[save_df['name'].str.contains('NC_045512.2'), 'date'] = '2019-12-26' #date of hospital admittance, not necessarily sequencing day
save_df['name'] = save_df['name'].str.replace('/', '_')
print(save_df)

save_df.to_csv('../data/COG-UK-NORW/NORW_treetime_dates.csv', sep=',', index=False)
```

**TimeTree with geographical locations (mugration model)**

First thing is to create a .csv of states, with the sample name as before and then a state, eg. `region`.

```python
! head -2 '../data/COG-UK-NORW/NORW_metadata.csv'
```

```python
readfile = '../data/COG-UK-NORW/NORW_metadata.csv'
df = pd.read_csv(readfile, header=0, usecols=['sequence_name', 'central_sample_id'])
df = df.rename(columns={'sequence_name':'name','central_sample_id':'region'}) #headers required by treetime
df['region'] = df['region'].str.split('-').str[0]
df['country'] = df['name'].str.split('/').str[0]
print(df)

#include only the names that match the names in the alignment
leaf_df = pd.read_csv('../data/COG-UK-NORW/NORW_200_leaftips.csv', header=None)
leaf_df = leaf_df.rename(columns={0:'name'})

#add dates to leaf_df, and format the names like in the tree
save_df = leaf_df.merge(df, how='left')
save_df['name'] = save_df['name'].str.replace('/', '_')
print(save_df)

save_df.to_csv('../data/COG-UK-NORW/NORW_treetime_states.csv', sep=',', index=False)
```

```python
#sandbox 1

norw_df = pd.read_csv("NORW_all.csv", usecols=['sequence_name','sample_date'], parse_dates=['sample_date'])
norw_df.set_index('sample_date', inplace=True)

norw_df.resample('1M') #bin is 1 month
print(norw_df.resample('1M'))
```

```python
#sandbox 2

df = pd.read_csv("NORW_all.csv", usecols=['sequence_name','sample_date'], nrows=20, parse_dates=['sample_date'])
df['bin_number'] = df.groupby(pd.Grouper(key='sample_date', freq='1M')).ngroup() #this is just here to make it easy to see what's going on: it isn't necessary
print(df)

df.groupby(pd.Grouper(key='sample_date', freq='1M')).sample(n=2, random_state=1)['sequence_name']

#now randomly sample up to n sequences from each bin
```

```python
#make this into a tidier function
#csv: path to metadata csv file, bin_size: time range for each bin (eg. '2D', '1W', '2M'), num_per_bin: how many samples to take from each time range (int)
def get_samples_by_date(csv, bin_size, num_per_bin):
    df = pd.read_csv(csv, usecols=['sequence_name','sample_date'], parse_dates=['sample_date'])
    bins = df.groupby(pd.Grouper(key='sample_date', freq=bin_size))
    smallest_bin = bins.size().min()
    sample_seqnames = bins.sample(n=min(smallest_bin, num_per_bin), random_state=1)['sequence_name']
    sample_seqnames = sample_seqnames.reset_index(drop=True) #reindex from 0
    print(f">{min(smallest_bin, num_per_bin)} seqs sampled per '{bin_size}' date range")
    return sample_seqnames #this is a series of sequence_names
```

```python
subseq = get_samples_by_date("NORW_all.csv", '1M', 3)
print(subseq)
subseq.to_csv("subseq.txt", index=False, header=None)
```
