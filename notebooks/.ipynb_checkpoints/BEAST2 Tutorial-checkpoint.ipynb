The point of `beauti2` is to generate a configuration file in `.xml` format that will be the input to `BEAST`.  The `.xml` file must contain:

* The data (typically a sequence alignment)
* The model specification
* Initial values and parameter constraints
* Settings of the MCMC algorithm
* Output options


BEAST2 uses a discretized gamma site model to account for variation in the rates at which different sites in the genome mutate.

This particular alignment file is discretized into four partitions: noncoding, codon position 1, cp2, and cp3.  The default method is for BEAST to create a separate tree and model for each of those partitions **Question: why/what is the biological motivation for doing this?**.  For this example, you're going to have it use all the partitions to create one tree.  To do this, select all 4 partitions and click "Link Trees" and "Link Clock Models".  Then, rename the trees and clock models to something more generic.

The 'tip dates' panel gets filled in if your samples are dated.  The example file sequences are all taken at the same time, so skip it this time.

Next is Site Model (even though we're producing a single tree, each partition gets its own site model, with different parameters estimated for each), Clock Model, and Priors.  The priors show your prior knowledge of what the model should be: more on this later.
