# Removing Outlier Sequences

How to use ``keep.tips`` (``drop.tips``) and ``treetime clock`` to remove outliers from a tree.

# Setup

```r
library(phangorn)
library(ggtreeExtra)
library(ggtree)
library(ggnewscale)
library(colorspace);
library(RColorBrewer);
library(dplyr)
library(tidyr)
library(ggplot2)
library (ggpubr)
library (stringr)
library(castor)
library(lubridate)
```

Define functions

```r
plot_tree_by_factor <- function(treefile, metadata, factor, treetype, mrsd){

# read in everything
tbl0 <- read.csv (metadata, sep = ",") 
tbl0[is.na(tbl0)] <- ""
if(treetype=="tree") {tre <- treefile} #already read into ape
else if(treetype=="newick") {tre <- read.tree (treefile)}
else if (treetype=="nexus") {tre <- read.nexus (treefile)}
traits <- tbl0 %>% slice(match(tre$tip.label, sequence_name)) #this line means you could use the entire cog_uk metadata
rownames(traits) <- traits$sequence_name
# traits$code <- substr(traits$central_sample_id,1,4)

# get plot inputs
groupInfo <- split (tre$tip.label, traits[factor])
grouptre <- groupOTU(tre, groupInfo, group_name="group1")
options(repr.plot.width=26, repr.plot.height=20) ; par(oma=c(0,0,0,0)) 
#getPal_1 = colorRampPalette( brewer.pal(8,"Dark2"))(length(unique(traits[,factor]))+1)  ## RdYlGn Set2
getPal_1 = getPal_lineage

# make plot
p <- ggtree (grouptre,  aes(color=group1),size=0.4, layout="rectangular") #, mrsd=mrsd)
p <- p + geom_tippoint(size=3, shape=21, aes(fill=group1))
p <- p + scale_color_manual(values = getPal_1, name="lineage") + scale_fill_manual(values = getPal_1, name="lineage")
p <- p + theme_tree2(text = element_text(size=30)) 
p <- p + theme(legend.position="left", legend.title = element_text(size=30), legend.text=element_text(size=30), legend.key.size = unit(1, 'cm'))
p
return(p)
}
```

Load tree + metadata; prepare standard lineage color palette

```r
treefile <- "NORW_Jan1toJun30_2020/full_timeframe/reformatted.aln.treefile"
metadata_file <- "NORW_Jan1toJun30_2020/full_timeframe/metadata_reformatted.csv"
```

```r
tre <- read.tree(treefile)
meta <- read.csv(metadata_file, sep=',')

tre <- midpoint(tre) #reroot
tre$edge.length <- tre$edge.length * 29903 #rescale to expected number of snps
```

```r
unique_lineages <- sort(unique(meta[["lineage"]]))
getPal_lineage = colorRampPalette(brewer.pal(8,"Dark2"))(length(unique_lineages) + 1)  ## RdYlGn Set2
names(getPal_lineage) = unique_lineages
```

# Original tree

Here's the original tree, colored by lineage:

```r
plot_tree_by_factor(tre, metadata_file, "lineage", "tree", "2020-06-30")
```

# Removing outliers with `keep.tips`

Plot the edge lengths to see where to draw the line:

```r
thresh = 5

tiplengths <- data.frame(tre$edge.length[1:Ntip(tre)])
names(tiplengths)[names(tiplengths) == 'tre.edge.length.1.Ntip.tre..'] <- 'tip_length'

options(repr.plot.width=10, repr.plot.height=8) ; par(oma=c(0,0,0,0)) 
p <- ggplot(tiplengths, aes(x=tip_length)) + geom_histogram(fill='skyblue3') + theme(text = element_text(size=25))
p <- p + geom_vline(xintercept = thresh, linetype = "dashed", colour = "red", size=2)
p
```

Get a list of all the tree tips for which edge length <- 5 expected SNPs:

```r
keep_tips <- tre$tip.label[(tre$edge.length[1:Ntip(tre)]) < 5]

length(keep_tips)
Ntip(tre)
```

Trim the outliers and plot the result:

```r
trimmed_tre <- keep.tip(tre, keep_tips)
Ntip(trimmed_tre)

plot_tree_by_factor(trimmed_tre, metadata_file, "lineage", "tree", "2020-06-30")
```
