```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('../cog_global_2022-02-06_public.csv.xz', parse_dates=['sample_date'], usecols=['sequence_name', 'sample_date'])
df['country'] = df['sequence_name'].str.split('/').str[0]
df['sequencing_centre'] = df['sequence_name'].str.split('/').str[1].str.split('-').str[0]

mask = (df['country'] == 'England')
df = df.loc[mask]
```

```python
df[df['sequence_name'].str.contains("PAHT")]
```

```python
sequencing_centres = pd.unique(df['sequencing_centre'])
print(sequencing_centres)

df['sequencing_centre'] = df['sequencing_centre'].str.slice(stop=4) #assume 'PAHTXXX' is a mistype, change to 'PAHT'

sequencing_centres = pd.unique(df['sequencing_centre'])
print(sequencing_centres)

#get earliest and latest dates
first_date = df['sample_date'].dt.date.min()
last_date = df['sample_date'].dt.date.max()

#df.set_index('sample_date', inplace=True)
```


look at this useful link: https://stackoverflow.com/questions/32012012/pandas-resample-timeseries-with-groupby

```python
df.set_index('sample_date', inplace=True)
grouper = df.groupby([pd.Grouper(freq='1M'), 'sequencing_centre'])
result = grouper['sequence_name'].count().unstack('sequencing_centre').fillna(0).astype('int64')
```

```python
df.groupby(pd.Grouper(freq='1M')).describe()
#.apply(lambda x: x.sample(min(2,len(x)))).reset_index(drop=True)
```

```python
result.to_csv("sequencing_centre_counts.csv")
```

```python
#make stacked bar plot out of dataframe
plt.figure()
result.plot.bar(figsize=(30,50), stacked=True)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.show()
```
